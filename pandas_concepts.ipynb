# How to apply multiple aggregations on a dataframe after applying groupby

```python
# aggregation as a dictionary
dic1 = {"age":"mean", "fare":"median"}
dic2 = {"age":"mean", "fare":"max"}
```

```python
print(dic1)
print(dic2)
```

```python
titanic_data.groupby(["embark_town", "who"]).agg(dic1).reset_index()
```

```python
titanic_data.groupby(["embark_town", "who"]).agg(dic2).reset_index()
```

```python
dic3 = {"age" :["mean","min","max"], "fare": ["min", "max", "sum","mean"]}
```

```python
titanic_data.groupby(["embark_town", "who"]).agg(dic3).reset_index()
```
